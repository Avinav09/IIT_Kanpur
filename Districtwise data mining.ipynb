```python
import requests
import json
import pandas as pd
```

```python
api_add = "https://api.covid19india.org/state_district_wise.json"
```

```python
json_data = requests.get(api_add).json()
#type(json_data)
json_data
```

```python
!pip install flat-table
```

```python
import flat_table
```

```python
df = pd.DataFrame()
```

```python
def district_data(df,statename):
    
    for features in json_data:
        if features == statename:
            df_x = pd.DataFrame(json_data[features]['districtData']).T
            df_c = flat_table.normalize(df_x)
            df_c['StateCode'] = json_data[features]['statecode']
            df_c['StateName'] = statename
            df = pd.concat([df,df_c], axis=0)
    return df
    
    
```

```python
for features in json_data:
    df = district_data(df,features)
```
