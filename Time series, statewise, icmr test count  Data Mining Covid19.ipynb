```python
import requests
import json
import pandas as pd
```

```python
api_add = "https://api.covid19india.org/data.json"
```

```python
json_data = requests.get(api_add).json()
```

```python
json_data
```

```python
case_time_series = json_data['cases_time_series']
```

```python
case_time_series_df = pd.DataFrame.from_dict(case_time_series)
case_time_series_df
```

```python
case_time_series_df = case_time_series_df[['date','dateymd','dailyconfirmed','dailydeceased','dailyrecovered','totalconfirmed','totaldeceased','totalrecovered']]
```

```python
case_time_series_df
```

```python
import sqlite3
```

```python
conn = sqlite3.connect(r"C:\Users\avina\OneDrive\Desktop\COVID 19 Data\DB Browser SQLite\COVID19.db")
```

```python
case_time_series_df.to_sql("case_time_series", conn, if_exists = "replace")
```

```python
statewise_data = json_data['statewise']
```

```python
statewise_df = pd.DataFrame.from_dict(statewise_data)
statewise_df
```

```python
statewise_df.columns
```
